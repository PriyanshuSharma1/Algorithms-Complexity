```python
# Function to generate a random array of given length
import time


def generate_random_array(length):
    return [random.randint(1, 10000) for i in range(length)]

# Function to generate a sorted array (best case) of given length
def generate_sorted_array(length):
    return [i for i in range(1, length + 1)]

# Function to generate a reverse sorted array (worst case) of given length
def generate_reverse_sorted_array(length):
    return [i for i in range(length, 0, -1)]

# Function to measure the execution time of a sorting algorithm
def measure_execution_time(sort_func, arr):
    start_time = time.time()
    sorted_arr = sort_func(arr)
    end_time = time.time()
    execution_time = end_time - start_time
    return execution_time

```

```python
# Generate random inputs
input_sizes = [10, 100, 1000, 10000, 100000, 1000000]  # Add more sizes as needed
execution_times_merge_avg = []
execution_times_quick_avg = []
execution_times_merge_best = []
execution_times_quick_best = []
execution_times_merge_worst = []
execution_times_quick_worst = []
```

```python
for size in input_sizes:
        arr=generate_random_array(size)
        arr1=generate_sorted_array(size)
        arr2=generate_reverse_sorted_array(size)
        execution_time_merge_avg = measure_execution_time(merge_sort, arr)
        execution_time_quick_avg = measure_execution_time(quick_sort, arr)
        execution_time_merge_best = measure_execution_time(merge_sort, arr1)
        execution_time_quick_best= measure_execution_time(quick_sort, arr1)
        execution_time_merge_worst = measure_execution_time(merge_sort, arr2)
        execution_time_quick_worst= measure_execution_time(quick_sort, arr2)
        execution_times_merge_avg.append(execution_time_merge_avg)
        execution_times_quick_avg.append(execution_time_quick_avg)
        execution_times_merge_best.append(execution_time_merge_best)
        execution_times_quick_best.append(execution_time_quick_best)
        execution_times_merge_worst.append(execution_time_merge_worst)
        execution_times_quick_worst.append(execution_time_quick_worst)
```

```python
 # Plot input size vs execution time graph
plt.plot(input_sizes, execution_times_merge_avg, label='Merge Avg Sort')
plt.plot(input_sizes, execution_times_quick_avg, label='Quick Avg Sort')
plt.plot(input_sizes, execution_times_merge_best, label='Merge Best Sort')
plt.plot(input_sizes, execution_times_quick_best, label='Quick Best Sort')
plt.plot(input_sizes, execution_times_merge_worst, label='Merge Worst Sort')
plt.plot(input_sizes, execution_times_quick_worst, label='Quick Worst Sort')
plt.xlabel('Input Size')
plt.ylabel('Execution Time (seconds)')
plt.title('Input Size vs Execution Time')
plt.legend()
plt.show()
```
