```python
import random
import matplotlib.pyplot as plt
from mergesortt import merge_sort
from quicksort import quick_sort
from generatearray import generate_random_array, measure_execution_time, generate_reverse_sorted_array,generate_sorted_array
from test import *
import time
import pandas as pd
```

```python
arr=[3,4,32,45,2,12,45,6,77]
print(merge_sort(arr))
```

```python
print(quick_sort(arr))
```

```python
average_case_merge_sort = []
best_case_merge_sort = []
worst_case_merge_sort = []
average_case_quick_sort = []
best_case_quick_sort = []
worst_case_quick_sort = []
```

```python
# random value
for n in range(1000,80000,1000):
    arr = [random.randint(0, 100000) for _ in range(n)]
    start = time.time()
    merge_sort(arr.copy())
    end = time.time()
    average_case_merge_sort.append([n, end - start])

    start = time.time()
    quick_sort(arr.copy())
    end = time.time()
    average_case_quick_sort.append([n, end - start])
# Convert lists to pandas DataFrames
df_merge_sort = pd.DataFrame(average_case_merge_sort, columns=['Input Size', 'Merge Sort Time'])
df_quick_sort = pd.DataFrame(average_case_quick_sort, columns=['Input Size', 'Quick Sort Time'])

# Merge the DataFrames on 'Input Size'
df_results = pd.merge(df_merge_sort, df_quick_sort, on='Input Size')

# Display the table
print("Average Case")
print(df_results)
```

```python

# best case and worst case test graph for merge sort
for n in range(1000, 80000,1000):
    arr = [i for i in range(n)]
    
    start = time.time()
    merge_sort(arr.copy())
    end = time.time()
    best_case_merge_sort.append([n, end - start])
    start = time.time()
    quick_sort(arr.copy())
    end = time.time()
    best_case_quick_sort.append([n, end - start])

    arr = [i for i in range(n, 0, -1)]
    
    start = time.time()
    merge_sort(arr.copy())
    end = time.time()
    worst_case_merge_sort.append([n, end - start])

    start = time.time()
    quick_sort(arr.copy())
    end = time.time()
    worst_case_quick_sort.append([n, end - start])
# Convert lists to pandas DataFrames
df_merge_sort = pd.DataFrame(worst_case_merge_sort, columns=['Input Size', 'Merge Sort Time'])
df_quick_sort = pd.DataFrame(worst_case_quick_sort, columns=['Input Size', 'Quick Sort Time'])

# Merge the DataFrames on 'Input Size'
df_results = pd.merge(df_merge_sort, df_quick_sort, on='Input Size')

# Display the table
print("Worst Case")
print(df_results)
```

```python
df1 = pd.DataFrame(average_case_merge_sort, columns=['n', 'time'])
df2 = pd.DataFrame(best_case_merge_sort, columns=['n', 'time'])
df3 = pd.DataFrame(worst_case_merge_sort, columns=['n', 'time'])

df4 = pd.DataFrame(average_case_quick_sort, columns=['n', 'time'])
df5 = pd.DataFrame(best_case_quick_sort, columns=['n', 'time'])
df6 = pd.DataFrame(worst_case_quick_sort, columns=['n', 'time'])
```

```python
plt.title("Merge Sort")
plt.xlabel("Number of data")
plt.ylabel("Time in seconds")
plt.plot(df1['n'], df1['time'], label="Average", linestyle='dotted')
plt.plot(df2['n'], df2['time'], label="Best", linestyle='solid')
plt.plot(df3['n'], df3['time'], label="Worst", linestyle='dashed')
plt.legend()
plt.show()
```

```python
plt.title("Quick Sort")
plt.xlabel("Number of data")
plt.ylabel("Time in seconds")
plt.plot(df4['n'], df4['time'], label="Average", linestyle='dotted')
plt.plot(df5['n'], df5['time'], label="Best", linestyle='solid')
plt.plot(df6['n'], df6['time'], label="Worst", linestyle='dashed')
plt.legend()
plt.show()
```
